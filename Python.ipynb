## **1. Compiler Vs Interpretor**
**Compiler vs. Interpreter** refers to two different approaches for executing code in programming languages. Here’s a comparison between the two:

### A. **Compiler:**
- **Definition:** A compiler translates the entire source code of a program into machine code or an intermediate code before execution.
- **Execution:** The whole program is compiled at once into an executable file, which is then run.
- **Output:** Produces an intermediate or binary executable file (e.g., `.exe` on Windows).
- **Performance:** Typically faster at runtime because the translation to machine code is done beforehand.
- **Errors:** Detects and reports errors for the entire program at compile time.
- **Examples of compiled languages:**
  - C, C++, Rust, Go, Swift.

### B. **Interpreter:**
- **Definition:** An interpreter translates code into machine language line-by-line and executes it immediately.
- **Execution:** The code is interpreted and executed line-by-line or statement-by-statement.
- **Output:** No separate output file is created; the interpreter runs the code directly.
- **Performance:** Slower at runtime since each line needs to be translated before execution.
- **Errors:** Errors are reported as soon as they occur, i.e., during execution.
- **Examples of interpreted languages:**
  - Python, Ruby, JavaScript, PHP.

### Key Differences:
| **Aspect**         | **Compiler**                        | **Interpreter**                   |
|--------------------|-------------------------------------|-----------------------------------|
| **Translation**    | Translates entire code at once      | Translates line-by-line           |
| **Execution**      | Executes compiled code separately   | Executes code directly            |
| **Runtime speed**  | Faster                              | Slower                           |
| **Error detection**| Reports errors after compilation    | Stops at the first error during execution |
| **Output**         | Produces an executable file         | Does not produce an intermediate file |
| **Memory usage**   | Requires more memory upfront        | Uses less memory upfront          |

Some languages, like Java, use both a **compiler and an interpreter**: Java source code is first compiled into bytecode (via a compiler), and the bytecode is then interpreted by the Java Virtual Machine (JVM).

## **2. PIP**
**PIP** stands for **"Pip Installs Packages"** and is the default package manager for **Python**. It is used to install and manage software packages or libraries written in Python. PIP allows users to download and install Python packages from the **Python Package Index (PyPI)**, a large repository of open-source Python software.

### Key Features of PIP:
1. **Installation of packages**: 
   - PIP can install packages from PyPI or other sources.
   - Example: 
     ```bash
     pip install <package_name>
     ```
     This command downloads and installs the package, including its dependencies.
   
2. **Uninstalling packages**:
   - PIP can also uninstall packages.
   - Example: 
     ```bash
     pip uninstall <package_name>
     ```

3. **List installed packages**:
   - PIP can display a list of all installed Python packages.
   - Example:
     ```bash
     pip list
     ```

4. **Upgrade packages**:
   - PIP can upgrade installed packages to their latest versions.
   - Example:
     ```bash
     pip install --upgrade <package_name>
     ```

5. **Dependency management**:
   - PIP handles package dependencies, meaning if a package depends on other libraries, PIP will download and install them automatically.

6. **Requirements file**:
   - You can use a requirements file to specify multiple packages to install at once, typically used in projects.
   - Example:
     ```bash
     pip install -r requirements.txt
     

PIP makes it easy to manage external libraries and packages in Python projects, significantly enhancing development efficiency.

## **5. Python Virtual Machine**
The **Python Virtual Machine (PVM)** is a crucial component in the process of executing Python code. It is part of the Python interpreter and acts as an intermediate layer that runs bytecode, allowing Python to be an interpreted language. The PVM is responsible for executing the compiled bytecode on the host machine.

Here’s a breakdown of how the PVM fits into the execution of Python programs:

### Key Concepts Related to Python Virtual Machine (PVM):

#### 1. **Python Source Code Execution Process**:
   The process of running a Python program involves several steps:
   - **Step 1**: You write Python source code in a `.py` file.
   - **Step 2**: The Python interpreter compiles the source code into **bytecode** (a lower-level, platform-independent representation). The bytecode is stored in `.pyc` files, typically within the `__pycache__` directory.
   - **Step 3**: The **PVM** takes this bytecode and interprets it, converting the bytecode instructions into machine code that the underlying hardware can execute.

#### 2. **Bytecode**:
   - Python does not directly execute the source code on the hardware. Instead, the code is first converted into **bytecode** (an intermediate representation that is platform-independent).
   - Bytecode is a set of instructions that the Python Virtual Machine can understand and execute.

#### 3. **Role of PVM**:
   - The **PVM** reads the bytecode, interprets it, and converts it into machine-level instructions that the operating system and hardware can run.
   - It operates as a **stack-based** machine that executes bytecode instruction by instruction.
   - PVM handles tasks like managing memory, variables, and flow control based on the bytecode.

#### 4. **Why PVM is Important**:
   - The PVM allows Python to be platform-independent. As long as the PVM is available for a particular platform (Windows, macOS, Linux, etc.), Python bytecode can be executed on that platform without any modification.
   - This is why Python is referred to as a **“portable language”**—the same Python code can run on multiple platforms with the help of the PVM.

#### 5. **PVM and Performance**:
   - The PVM is **slower than direct machine code execution** because it is an additional layer of interpretation. However, tools like **PyPy** (an alternative Python implementation) attempt to improve performance by using Just-In-Time (JIT) compilation, which translates parts of the bytecode into machine code on the fly, reducing the need for repeated interpretation.

#### 6. **Garbage Collection**:
   - The PVM also handles garbage collection—automatically freeing up memory by removing objects that are no longer in use.

### Python Code Execution Flow:
1. **Source Code** → (Python Compiler) → **Bytecode** → (PVM) → **Machine Code**.

### Diagram Representation:
```
                                    Python Source Code (.py)
                                          |
                                          V
                                     Python Compiler
                                          |
                                          V
                                     Bytecode (.pyc)
                                          |
                                          V
                                   Python Virtual Machine (PVM)
                                          |
                                          V
                                   Machine Code (Hardware Execution)
```



### Advantages of PVM:
- **Portability**: Since the PVM interprets bytecode rather than machine code, Python programs can run on any platform that has a Python interpreter.
- **Simplifies Development**: The PVM handles platform-specific issues, so developers don’t need to worry about low-level details of the operating system or hardware.
  
In summary, the **Python Virtual Machine (PVM)** is an essential component that interprets Python bytecode and converts it into executable machine code, enabling Python programs to run on different platforms without modification.

## **6. Memory Management in Python**
### Key Components of Python Memory Management:

1. **Memory Allocation**:
   - **Stack Memory**: Used for function calls and local variables, automatically managed by the system.
   - **Heap Memory**: Stores objects like lists, classes, and dictionaries. Dynamically managed by Python’s memory manager.

2. **Reference Counting**:
   - Each object has a **reference count**. The count increases when the object is referenced and decreases when the reference is deleted or reassigned.
   - When the reference count drops to **zero**, the memory is automatically deallocated.

### Best Practices for Memory Optimization:

1. **Choose Efficient Data Structures**: Select appropriate data structures and algorithms to minimize memory usage.
2. **Avoid Unnecessary Copies**: Work with object references instead of duplicating data.
3. **Use Generators**: Generate values on-the-fly for memory-efficient iteration, especially with large datasets.
4. **Optimize Loops**: Use list comprehensions or generator expressions to reduce memory overhead in loops.
5. **Use Context Managers**: Manage resources like files efficiently using `with` statements to ensure timely release.
6. **Profile Memory Usage**: Use tools to identify memory bottlenecks and optimize critical sections.

### **3. Operator Precedence**

Determines the order in which operators are evaluated.  
**Precedence** (highest to lowest):
1. Parentheses `()`
2. Exponentiation `**`
3. Unary `+`, `-`
4. Multiplication, Division `*`, `/`, `//`, `%`
5. Addition, Subtraction `+`, `-`
6. Relational `<`, `<=`, `>`, `>=`, `==`, `!=`
7. Logical `and`, `or`, `not`

Example:
```python
x = 2 + 3 * 4  # Multiplication is done first
print(x)  # 14

y = (2 + 3) * 4  # Parentheses take precedence
print(y)  # 20
```

---

### **4. Identity Operators**

Used to compare memory locations of two objects.  
**Operators**: `is`, `is not`

Example:
```python
x = [1, 2, 3]
y = [1, 2, 3]
z = x

print(x is y)       # False, because x and y are different objects
print(x is z)       # True, because z references x
print(x is not y)   # True
```

## **The difference between `for` and `while` loops lies in how they are structured and their common use cases.** <br> Here's a breakdown with examples:

### 1. **`for` Loop**
- **Purpose**: Typically used when the number of iterations is known beforehand or when iterating over a sequence (like a list, range, or string).
- **Structure**:
  - Executes a block of code a specific number of times.
  - Automatically handles the loop counter or iterator.

#### Example: Print numbers from 1 to 5
```python
for i in range(1, 6):
    print(i)
```
**Explanation**:  
- The loop iterates over a range of numbers (1 to 5).
- `i` automatically takes each value from the sequence.

---

### 2. **`while` Loop**
- **Purpose**: Used when the number of iterations is not known in advance and depends on a condition.
- **Structure**:
  - Repeats the code block as long as the condition is `True`.
  - The user must explicitly manage the loop counter (if needed).

#### Example: Print numbers from 1 to 5
```python
i = 1
while i <= 5:
    print(i)
    i += 1  # Increment the counter to avoid infinite loop
```
**Explanation**:  
- The loop continues as long as the condition `i <= 5` is `True`.
- The programmer manually increments `i`.

---

### Key Differences

| Feature          | `for` Loop                          | `while` Loop                            |
|-------------------|-------------------------------------|-----------------------------------------|
| **Iterations**    | Known beforehand                   | Unknown; based on a condition           |
| **Use Case**      | Iterating over sequences, ranges    | Condition-based loops                   |
| **Counter Mgmt**  | Automatic                          | Manual                                  |
| **Termination**   | Stops when range/sequence ends     | Stops when condition becomes `False`    |

---

### Example Comparison: Factorial Calculation
#### Using `for` Loop:
```python
num = 5
factorial = 1
for i in range(1, num + 1):
    factorial *= i
print("Factorial using for loop:", factorial)
```

#### Using `while` Loop:
```python
num = 5
factorial = 1
i = 1
while i <= num:
    factorial *= i
    i += 1
print("Factorial using while loop:", factorial)
```

Both achieve the same result but show different scenarios where `for` or `while` might be preferred!

### **1. Difference Between a Function and a Method**

- **Function**: A function is a block of reusable code that can take input and return a result. It is independent and can be called in a standalone manner.

- **Method**: A method is a function that is associated with an object. It is defined within a class and operates on instances of that class.

**Example:**

```python
# Function
def greet(name):
    return f"Hello, {name}"

print(greet("Alice"))  # Output: Hello, Alice

# Method
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}"

p = Person("Alice")
print(p.greet())  # Output: Hello, Alice
```

---

### **2. Defining a Function**

A function in Python is defined using the `def` keyword followed by the function name, parameters in parentheses, and a colon. The body of the function is indented.

**Example:**

```python

    return a + b
```

## **2. Function Decorators**
Decorators in Python are **higher-order functions** that modify another function **without modifying its actual code**.

### **Basic Decorator Example**
```python
def decorator_function(original_function):
    def wrapper_function():
        print("Before the function call")
        original_function()
        print("After the function call")
    return wrapper_function

@decorator_function
def display():
    print("Hello, World!")

display()
```

### **Output:**
```
Before the function call
Hello, World!
After the function call
```

---

### **3. Property Decorators (@property)**
Property decorators allow us to define **getter, setter, and deleter methods** in an object-oriented way.

```python
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):  # Getter method
        return self._name

    @name.setter
    def name(self, new_name):  # Setter method
        if len(new_name) > 2:
            self._name = new_name
        else:
            print("Name too short!")

p = Person("Alice")
print(p.name)  # Output: Alice

p.name = "Bo"  # Name too short!
p.name = "Bob"
print(p.name)  # Output: Bob
```

## **4. Generators**
Generators are **iterators that yield values lazily** using the `yield` keyword instead of `return`. They are **memory efficient**.

### **Basic Generator Example**
```python

    yield 1
    yield 2
    yield 3

gen = my_generator()

print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3
```

### **5. Iterators in Python**
An **iterator** is an object that **implements `__iter__()` and `__next__()` methods**.

### **Custom Iterator**
```python
class MyNumbers:
    def __init__(self, start, end):
        self.start = start
        self.end = end

    def __iter__(self):
        self.current = self.start
        return self

    def __next__(self):
        if self.current > self.end:
            raise StopIteration
        self.current += 1
        return self.current - 1

numbers = MyNumbers(1, 5)
for num in numbers:
    print(num)
```

# **8.1. String Slicing**



### **String Slicing Syntax:**

```python
string[start:end:step]
```

- **start**: The index at which the slice begins (inclusive). Defaults to `0` if not provided.
- **end**: The index at which the slice ends (exclusive). Defaults to the end of the string if not provided.
- **step**: The step size, i.e., how many characters to skip between slices. Defaults to `1` if not provided.

### **Indexing in Python Strings:**
- String indices start from `0` for the first character.
- Negative indices can be used to count from the end, where `-1` is the last character, `-2` is the second last character, and so on.

---

### **1. Basic Slicing (Extracting a Substring)**

**Example:**

```python
text = "Hello, World!"

# Extract characters from index 0 to 4 (exclusive)
substring = text[0:5]
print(substring)  # Output: Hello
```

**Explanation:**  
- The slice starts at index `0` (inclusive) and ends at index `5` (exclusive), extracting the characters `"Hello"`.

---

### **2. Omitting the Start and End (Default Behavior)**

If you omit the `start` and `end`, the slice will include the entire string.

**Example:**

```python
text = "Hello, World!"

# Extract the entire string
substring = text[:]
print(substring)  # Output: Hello, World!
```

**Explanation:**  
- By leaving out both `start` and `end`, we get the entire string.

---

### **3. Slicing with Negative Indices**

Negative indices count from the end of the string.

**Example:**

```python
text = "Hello, World!"

# Extract the last 6 characters
substring = text[-6:]
print(substring)  # Output: World!
```

**Explanation:**  
- The slice starts at index `-6`, which is the `W`, and goes until the end of the string.

---

### **4. Using Step to Skip Characters**

The `step` value can be used to skip characters.

**Example:**

```python
text = "Hello, World!"

# Extract every second character from index 0 to 11
substring = text[0:12:2]
print(substring)  # Output: Hoo ol
```

**Explanation:**  
- The slice starts at index `0` and goes up to index `12` (exclusive). It skips every second character, resulting in `"Hoo ol"`.

---

### **5. Reversing a String Using Negative Step**

You can reverse a string by using a negative `step`.

**Example:**

```python
text = "Hello, World!"

# Reverse the string
reversed_text = text[::-1]
print(reversed_text)  # Output: !dlroW ,olleH
```

**Explanation:**  
- By using `[::-1]`, we create a slice with no `start` or `end`, but with a `step` of `-1`, which reverses the string.

---

### **6. Slicing with Only Start or End**

You can slice by specifying only the `start` or `end`, which will slice from that point to the end or start, respectively.

**Example 1: Only `start` specified:**

```python
text = "Hello, World!"

# Slice from index 7 to the end
substring = text[7:]
print(substring)  # Output: World!
```

**Explanation:**  
- The slice starts at index `7` and continues until the end of the string, resulting in `"World!"`.

**Example 2: Only `end` specified:**

```python
text = "Hello, World!"

# Slice from the beginning up to index 5
substring = text[:5]
print(substring)  # Output: Hello
```

**Explanation:**  
- The slice starts from the beginning (index `0`) and goes up to index `5` (exclusive), resulting in `"Hello"`.

---

### **7. Slicing with Steps and Negative Indices**

You can combine negative indices with a `step` to extract characters in reverse order.

**Example:**

```python
text = "Hello, World!"

# Slice characters from index -2 to index -7 in reverse
substring = text[-2:-8:-1]
print(substring)  # Output: ol, W
```

**Explanation:**  
- The slice starts at index `-2` (the `!` character) and goes backward to index `-8` (exclusive), extracting `"ol, W"` in reverse order.

---

### **8. Slicing with Both Positive and Negative Indices**

You can combine both positive and negative indices for slicing.

**Example:**

```python
text = "Hello, World!"

# Slice from index 0 to the second last character
substring = text[0:-1]
print(substring)  # Output: Hello, World
```

**Explanation:**  
- The slice starts at index `0` (inclusive) and ends at the second last character (index `-1`), giving `"Hello, World"`.

---

### **9. Slicing with Step > 1 (Skip Multiple Characters)**

You can skip more than one character by setting the step size to a value greater than 1.

**Example:**

```python
text = "abcdefghijklmnopqrstuvwxyz"

# Extract every third character
substring = text[::3]
print(substring)  # Output: adgjmpsvy
```

**Explanation:**  
- The slice starts at the beginning of the string and takes every third character, resulting in `"adgjmpsvy"`.

---

### **10. Slicing with Complex Patterns**

Slicing is very flexible, and you can create complex patterns using combinations of `start`, `end`, and `step`.

**Example:**

```python
text = "abcdefghi"

# Extract characters from index 1 to 7, every 2nd character
substring = text[1:8:2]
print(substring)  # Output: bdf
```

**Explanation:**  
- The slice starts at index `1`, ends at index `8` (exclusive), and takes every second character, resulting in `"bdf"`.

---

### **Summary of Key Concepts:**

- **start**: The index where the slice starts (inclusive).
- **end**: The index where the slice ends (exclusive).
- **step**: The number of characters to skip between each character in the slice.
- **Negative indices**: Count from the end of the string, where `-1` is the last character.
- **Default behavior**: Omitting `start`, `end`, or `step` uses default values (whole string, full range, and step of `1`).

# **10. Sets**

### **Introduction to Sets**

A **set** in Python is an unordered collection of unique elements. Unlike lists or tuples, sets do not allow duplicate values, and their elements are not indexed.

**Key Characteristics**:
- Unordered
- No duplicates
- Mutable (you can add/remove elements)

**Example**:
```python
# Creating a set
set1 = {1, 2, 3, 4}
set2 = set([5, 6, 7])
print(set1)  # Output: {1, 2, 3, 4}
print(set2)  # Output: {5, 6, 7}
```

---


### **Internal Working of Sets**

Sets in Python are implemented using **hash tables**:
- Elements in a set are hashed to determine their storage location.
- This hashing ensures **fast lookups** for membership checks (e.g., `x in set`).

**Note**:
- Since elements are hashed, only **hashable** objects (e.g., numbers, strings, tuples) can be added to a set. Mutable objects like lists cannot be part of a set.

---



### **Set in Mathematics**

Sets in Python follow basic mathematical set operations like **union**, **intersection**, and **difference**.

**Example**:
```python
A = {1, 2, 3}
B = {3, 4, 5}

# Union
print(A | B)  # {1, 2, 3, 4, 5}

# Intersection
print(A & B)  # {3}

# Difference
print(A - B)  # {1, 2}
```

---



### **Set Methods: Union, Intersection, and Difference**

1. **Union (`union()` or `|`)**  
   Combines all elements from two sets.
   ```python
   A = {1, 2, 3}
   B = {3, 4, 5}
   print(A.union(B))  # {1, 2, 3, 4, 5}
   print(A | B)       # {1, 2, 3, 4, 5}
   ```

2. **Intersection (`intersection()` or `&`)**  
   Finds common elements between sets.
   ```python
   print(A.intersection(B))  # {3}
   print(A & B)              # {3}
   ```

3. **Difference (`difference()` or `-`)**  
   Returns elements in one set but not in another.
   ```python
   print(A.difference(B))  # {1, 2}
   print(A - B)            # {1, 2}
   ```

4. **Symmetric Difference (`symmetric_difference()` or `^`)**  
   Returns elements not common to both sets.
   ```python
   print(A.symmetric_difference(B))  # {1, 2, 4, 5}
   print(A ^ B)                      # {1, 2, 4, 5}
   ```

---



### **Set Methods: Adding and Deleting**

1. **`add(element)`**  
   Adds a single element to the set.
   ```python
   s = {1, 2}
   s.add(3)
   print(s)  # {1, 2, 3}
   ```

2. **`update(iterable)`**  
   Adds multiple elements from an iterable to the set.
   ```python
   s = {1, 2}
   s.update([3, 4, 5])
   print(s)  # {1, 2, 3, 4, 5}
   ```

3. **`copy()`**  
   Creates a shallow copy of the set.
   ```python
   s = {1, 2, 3}
   s_copy = s.copy()
   print(s_copy)  # {1, 2, 3}
   ```

4. **`remove(element)`**  
   Removes an element. Raises an error if the element is not present.
   ```python
   s = {1, 2, 3}
   s.remove(2)
   print(s)  # {1, 3}
   # s.remove(5)  # Raises KeyError
   ```

5. **`discard(element)`**  
   Removes an element. Does not raise an error if the element is not present.
   ```python
   s = {1, 2, 3}
   s.discard(2)
   print(s)  # {1, 3}
   s.discard(5)  # No error
   ```

6. **`pop()`**  
   Removes and returns an arbitrary element from the set.
   ```python
   s = {1, 2, 3}
   element = s.pop()
   print(element)  # Output could be 1, 2, or 3
   print(s)  # Remaining elements
   ```

7. **`clear()`**  
   Removes all elements from the set.
   ```python
   s = {1, 2, 3}
   s.clear()
   print(s)  # set()
   ```

---



### **Set Comprehensions**

Set comprehensions provide a concise way to create sets.

**Example**:
```python

print(squares)  # {1, 4, 9, 16, 25}

# Filter even numbers from 1 to 10

```

### **Examples of Using Sets**

   ```python

   print(unique_numbers)  # {1, 2, 3, 4, 5}
   ```

2. **Checking Subset/Superset**:
   ```python

### **Dictionary Methods: Adding and Removing**

#### **Adding Methods**

   Creates a shallow copy of the dictionary.
   ```python

   copy_dict = original.copy()

2. **`update(iterable)`**
   Updates the dictionary with key-value pairs from another dictionary or iterable.
   ```python

   Inserts a key with a default value if the key does not exist.

   print(d)  # Output: {'b': 2}
   ```

   Removes and returns the last key-value pair as a tuple.
   ```python

   print(d)  # Output: {'a': 1}
   ```

3. **`clear()`**
   Removes all elements from the dictionary.
   ```python

# **11.1 Difference between List, Tuple, Sets and Dictionary**

Here’s a concise comparison of **lists**, **tuples**, **sets**, and **dictionaries** in Python:

### 1. **List**
- **Definition**: Ordered, mutable collection of elements.
- **Syntax**: `[1, 2, 3]`
- **Key Features**:
  - Allows duplicates.
  - Elements can be accessed by index.
  - Can hold heterogeneous data types (e.g., integers, strings, objects).

---




### 2. **Tuple**
- **Definition**: Ordered, immutable collection of elements.
- **Syntax**: `(1, 2, 3)`
- **Key Features**:
  - Allows duplicates.
  - Elements can be accessed by index.
  - Cannot be modified after creation (immutable).
  - Useful for fixed data (e.g., coordinates, configuration settings).

---



### 3. **Set**
- **Definition**: Unordered, mutable collection of unique elements.
- **Syntax**: `{1, 2, 3}`
- **Key Features**:
  - Does not allow duplicates.
  - No indexing or ordering.
  - Useful for membership testing and mathematical operations like union, intersection.

---



### 4. **Dictionary**
- **Definition**: Unordered, mutable collection of key-value pairs.
- **Syntax**: `{'key1': 'value1', 'key2': 'value2'}`
- **Key Features**:
  - Keys are unique; values can be duplicated.
  - Keys are used for indexing.
  - Efficient for lookups and data mapping.

---



### Summary Table

| Feature          | List               | Tuple              | Set               | Dictionary           |
|-------------------|--------------------|--------------------|-------------------|----------------------|
| **Order**         | Ordered           | Ordered           | Unordered         | Unordered            |
| **Mutable**       | Yes               | No                | Yes               | Yes                  |
| **Duplicates**    | Allowed           | Allowed           | Not allowed       | Keys: No, Values: Yes|

# **12. OOPS: Object oriented programming structures**

## **Object-Oriented Programming (OOP) in Python**
Object-Oriented Programming (OOP) is a programming paradigm based on the concept of **objects**, which contain data in the form of **attributes** (variables) and **methods** (functions). Python is an **object-oriented language**, which means it supports OOP principles such as:

### **Explanation**:

## **2. Encapsulation (Get & Set)**
Encapsulation restricts direct access to object data and allows controlled modification using **getter and setter methods**.

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private Variable

    def get_balance(self):  # Getter
        return self.__balance

    def set_balance(self, amount):  # Setter
        if amount >= 0:
            self.__balance = amount
        else:
            print("Invalid balance amount!")

# Object Creation
account = BankAccount(1000)

# Accessing Private Variable using Getter
print(account.get_balance())

# Modifying Private Variable using Setter
account.set_balance(2000)
print(account.get_balance())

account.set_balance(-500)  # Invalid update
```

### **Explanation**:
- `__balance` is a **private variable**, not accessible directly.
- `get_balance()` returns the balance.
- `set_balance(amount)` updates the balance only if it's valid.

---

## **3. Abstraction**
Abstraction hides complex implementation details and shows only the necessary parts.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):  # Abstract Class
    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        print("Car started with a key.")

class Bike(Vehicle):
    def start(self):
        print("Bike started with a self-start.")

# Object Creation
car = Car()
bike = Bike()
car.start()
bike.start()
```

### **Explanation**:
- `Vehicle` is an **abstract class** (cannot be instantiated).
- `start()` is an **abstract method**, forcing subclasses to implement it.
- `Car` and `Bike` provide their own implementations of `start()`.

- `__age` is a **private variable**, so it can't be accessed directly.

# **13. Inheritence and polymorphism**

Inheritance and polymorphism are key features of Object-Oriented Programming in Python. Let’s go over each concept with examples and explanations.

---

## **1. Inheritance in Python**
Inheritance allows one class (**child class**) to acquire the properties and behaviors of another class (**parent class**). This promotes code reusability.

### **Example**
```python
# Parent Class
class Animal:
    def __init__(self, name):
        self.name = name

    def sound(self):
        return "Some sound"

# Child Class (Inheriting Animal)
class Dog(Animal):
    def sound(self):  # Overriding parent method
        return "Bark"

# Creating Objects
dog = Dog("Buddy")
print(dog.name)        # Output: Buddy
print(dog.sound())     # Output: Bark
```

### **Explanation**
- `Animal` is the parent class with an attribute (`name`) and a method (`sound()`).
- `Dog` inherits from `Animal` and **overrides** `sound()`.
- The child class can use both its own methods and inherited ones.

---

## **2. Constructors in Inheritance**
When a child class is instantiated, its constructor (`__init__()`) is called. If the child class doesn't define a constructor, the parent class's constructor is used.

### **Example**
```python
class Parent:
    def __init__(self):
        print("Parent Constructor Called")

class Child(Parent):
    def __init__(self):
        print("Child Constructor Called")

c = Child()
```

### **Output**
```
Child Constructor Called
```

### **Explanation**
- The child class defines its own constructor, so the parent’s constructor is **not called**.

---

## **3. Overriding Superclass Constructors and Methods**
We can override the parent class's constructor and methods in the child class.

### **Example**
```python
class Parent:
    def __init__(self):
        print("Parent Constructor Called")

    def display(self):
        print("Parent Class Method")

class Child(Parent):
    def __init__(self):
        print("Child Constructor Called")

    def display(self):
        print("Child Class Method")

c = Child()
c.display()
```

### **Output**
```
Child Constructor Called
Child Class Method
```

### **Explanation**
- The `Child` class **overrides** the parent’s `display()` method.
- The child’s constructor **replaces** the parent’s constructor.

---

## **4. The `super()` Method**
The `super()` function allows access to parent class methods and constructor.

### **Example**
```python
class Parent:
    def __init__(self):
        print("Parent Constructor Called")

class Child(Parent):
    def __init__(self):
        super().__init__()  # Calls Parent Constructor
        print("Child Constructor Called")

c = Child()
```

### **Output**
```
Parent Constructor Called
Child Constructor Called
```

### **Explanation**
- `super().__init__()` calls the parent constructor before executing the child constructor.

## **6. Method Resolution Order (MRO)**
MRO defines the order in which methods are searched in multiple inheritance.

```python
class A:
    def show(self):
        print("A class method")

class B(A):
    pass

class C(A):
    def show(self):
        print("C class method")

class D(B, C):
    pass

obj = D()
obj.show()
```

### **Output**
```
C class method
```

### **Explanation**
- Python follows **C3 Linearization (Depth-First Left-to-Right)**.
- The order is: `D → B → C → A`.

# **15. Exceptions**


### **Errors in a Python Program**
Errors in Python can occur during the execution of a program. Errors can be categorized into two main types:
1. **Syntax Errors**: Occur when the code violates the Python syntax rules.
2. **Exceptions**: Errors detected during execution (runtime) that are not unconditionally fatal.

---

### **Exceptions**
An **exception** is an event that occurs during the execution of a program and disrupts its normal flow. Common exceptions include `ZeroDivisionError`, `ValueError`, `TypeError`, etc.

---

### **Exception Handling**
Exception handling ensures that a program does not crash and can handle errors gracefully. Python provides the `try-except` block for exception handling.

---

### **Types of Exceptions**
Here are some common exceptions in Python:
- **`ZeroDivisionError`**: Division by zero is not allowed.
- **`ValueError`**: Invalid value.
- **`KeyError`**: Accessing a non-existent key in a dictionary.
- **`TypeError`**: Unsupported operation between types.
- **`IndexError`**: Accessing an invalid index in a list.

---

### **The `try-except` Block**
The `try-except` block is used to catch and handle exceptions.

#### **Code Example: `try-except` Block**
```python
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print("Error: Division by zero is not allowed!", e)
finally:
    print("Execution completed.")
```
**Output**:
```
Error: Division by zero is not allowed! division by zero
Execution completed.
```

---


### **User-Defined Exceptions**
You can create custom exceptions by subclassing the `Exception` class.

#### **Code Example: User-Defined Exceptions**
```python
class MyCustomError(Exception):
    def __init__(self, message):
        self.message = message
        super().__init__(self.message)

def check_positive_number(value):
    if value < 0:
        raise MyCustomError("Negative numbers are not allowed!")
    print(f"{value} is a positive number.")

try:
    check_positive_number(-10)
except MyCustomError as e:
    print(f"Custom Error Caught: {e}")
finally:
    print("End of exception handling.")
```

**Output**:
```
Custom Error Caught: Negative numbers are not allowed!
End of exception handling.
```

---

### **Summary**
- Exceptions handle runtime errors.

# **17. Regular expressions in python**

### **Regular Expressions in Python (re module)**
Regular Expressions (**RegEx**) in Python allow you to search, match, and manipulate strings using patterns.

To use regular expressions in Python, you need to import the `re` module:
```python
import re
```

---

## **1. Basic Functions of `re` Module**
| Function | Description |
|----------|------------|
| `re.match()` | Checks if the pattern matches **only at the beginning** of the string. |
| `re.search()` | Searches for a pattern **anywhere in the string**. |
| `re.findall()` | Returns **all matches** in the string as a list. |
| `re.finditer()` | Returns an iterator of match objects. |
| `re.sub()` | Replaces matches with a new string. |
| `re.split()` | Splits the string by the matched pattern. |

### **Example: Using `search`, `match`, and `findall`**
```python
text = "Hello, my number is 12345 and my friend's number is 67890."

# Match at the beginning of the string
match_result = re.match(r"\d+", text)  
print(match_result)  # Output: None (because numbers are not at the start)

# Search for first occurrence
search_result = re.search(r"\d+", text)
print(search_result.group())  # Output: 12345

# Find all occurrences
findall_result = re.findall(r"\d+", text)
print(findall_result)  # Output: ['12345', '67890']
```

---

## **2. Sequence Characters in Regular Expressions**
Sequence characters help define **character classes**.

| Sequence | Meaning | Example |
|----------|---------|---------|
| `\d` | Matches any **digit (0-9)** | `re.findall(r"\d+", "12 apples, 34 bananas")` → `['12', '34']` |
| `\D` | Matches **non-digits** | `re.findall(r"\D+", "123abc")` → `['abc']` |
| `\w` | Matches **word characters** (`[a-zA-Z0-9_]`) | `re.findall(r"\w+", "Hello_123")` → `['Hello_123']` |
| `\W` | Matches **non-word characters** | `re.findall(r"\W+", "Hello! How are you?")` → `['! ', ' ']` |
| `\s` | Matches **whitespace (space, tab, newline)** | `re.findall(r"\s+", "Hello  World")` → `['  ']` |
| `\S` | Matches **non-whitespace** | `re.findall(r"\S+", "Hello  World")` → `['Hello', 'World']` |
| `.` | Matches **any character except newline** | `re.findall(r'H.llo', 'Hello Hallo Hullo')` → `['Hello', 'Hallo', 'Hullo']` |
| `^` | Matches **start of string** | `re.search(r"^Hello", "Hello World")` → `Hello` |
| `$` | Matches **end of string** | `re.search(r"World$", "Hello World")` → `World` |

### **Example: Using Sequence Characters**
```python
text = "Hello123 World!"

print(re.findall(r"\d", text))   # ['1', '2', '3']
print(re.findall(r"\D", text))   # ['H', 'e', 'l', 'l', 'o', ' ', 'W', 'o', 'r', 'l', 'd', '!']
print(re.findall(r"\s", text))   # [' ']
print(re.findall(r"\w", text))   # ['H', 'e', 'l', 'l', 'o', '1', '2', '3', 'W', 'o', 'r', 'l', 'd']
```

---

## **3. Quantifiers in Regular Expressions**
Quantifiers specify how many times a pattern should occur.

| Quantifier | Meaning | Example |
|------------|---------|---------|
| `*` | **0 or more times** | `re.findall(r'ab*', 'abb abbbb a')` → `['abb', 'abbbb', 'a']` |
| `+` | **1 or more times** | `re.findall(r'ab+', 'abb abbbb a')` → `['abb', 'abbbb']` |
| `?` | **0 or 1 time (optional)** | `re.findall(r'ab?', 'abb abbbb a')` → `['ab', 'ab', 'a']` |
| `{n}` | **Exactly `n` times** | `re.findall(r'\d{3}', '123 4567 89')` → `['123', '456']` |
| `{n,}` | **At least `n` times** | `re.findall(r'\d{2,}', '1 12 123')` → `['12', '123']` |
| `{n,m}` | **Between `n` and `m` times** | `re.findall(r'\d{2,4}', '1 12 123 12345')` → `['12', '123', '1234']` |

### **Example: Using Quantifiers**
```python
text = "aaaa abc abbb abb"
print(re.findall(r"ab*", text))  # ['a', 'ab', 'abb', 'abbb', 'ab']
print(re.findall(r"ab+", text))  # ['ab', 'abb', 'abbb']
print(re.findall(r"ab?", text))  # ['a', 'ab', 'ab', 'ab']
```

---

## **4. Special Characters in Regular Expressions**
| Special Character | Meaning | Example |
|-------------------|---------|---------|
| `\` | Escape character | `re.findall(r"\.com", "example.com")` → `['.com']` |
| `|` | OR operator | `re.findall(r'cat|dog', 'cat and dog')` → `['cat', 'dog']` |
| `()` | Grouping | `re.findall(r'(ab)+', 'ababab')` → `['ab', 'ab', 'ab']` |
| `[]` | Character set | `re.findall(r'[aeiou]', 'hello')` → `['e', 'o']` |
| `[^]` | Negation inside set | `re.findall(r'[^aeiou]', 'hello')` → `['h', 'l', 'l']` |

### **Example: Using Special Characters**
```python
text = "hello.com, example.com, test.net"
print(re.findall(r"\.com", text))  # ['.com', '.com']
print(re.findall(r"hello|test", text))  # ['hello', 'test']
print(re.findall(r"[aeiou]", text))  # ['e', 'o', 'o', 'e', 'a', 'e', 'o', 'e']
print(re.findall(r"[^aeiou]", text))  # All consonants and symbols
```

---

## **5. Substituting & Splitting Strings**
### **(a) Replace Text Using `re.sub()`**
```python

print(new_text)  # "My phone number is ***-***-****."
```

### **(b) Split a String Using `re.split()`**
```python

print(split_text)  # ['apple', 'banana', 'cherry', 'orange']
```

# **19. Matplotlib and Seaborn**

### **Overview of Matplotlib**

**Matplotlib** is a widely-used Python library for creating static, interactive, and animated visualizations. It is highly customizable and allows users to generate a variety of plots, from basic line charts to complex multi-axes figures.

---



#### **Key Features of Matplotlib**
1. **Wide Variety of Plots**: Line plots, bar plots, scatter plots, histograms, and more.
2. **Highly Customizable**: Control over every aspect of the figure, including size, colors, fonts, and annotations.
3. **Integration**: Works well with NumPy, pandas, and other scientific libraries.
4. **Low-Level Control**: Offers detailed control over plot elements, making it flexible but requiring more code.

---



### **Overview of Seaborn**

**Seaborn** is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Seaborn simplifies many tasks, such as working with DataFrames and applying aesthetic styles to plots.

---



#### **Key Features of Seaborn**
1. **Built-In Aesthetic Themes**: Automatically applies beautiful styling to plots.
2. **Simplified Plotting**: Requires less code than Matplotlib for complex plots.
3. **Integrated with pandas**: Works seamlessly with DataFrames.
4. **Statistical Visualizations**: Includes specialized plots like violin plots, pair plots, and heatmaps.

---



### **Differences Between Matplotlib and Seaborn**

| Feature                  | Matplotlib                          | Seaborn                                   |
|--------------------------|-------------------------------------|------------------------------------------|
| **Purpose**              | General-purpose plotting library    | Specialized for statistical data visualization |
| **Ease of Use**          | Requires more detailed coding       | Simplifies complex visualizations        |
| **Aesthetic Styling**    | Basic styling by default            | Comes with built-in themes and styles    |
| **Integration**          | Works with NumPy and pandas         | Designed specifically for pandas DataFrames |
| **Types of Plots**       | Basic plots like line, scatter, etc. | Advanced statistical plots (e.g., heatmaps, pair plots) |
| **Customization**        | Fully customizable but verbose      | Less customizable than Matplotlib        |

---


### Here’s a **side-by-side comparison** of **Matplotlib vs. Seaborn** for various charts:  

- **Line Plot**  
- **Scatter Plot**  
- **Bar Chart**  
- **Histogram**  
- **Pie Chart**  

---

### **1. Line Plot: Matplotlib vs. Seaborn**

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)
df = pd.DataFrame({"x": x, "y": y})

# Side-by-side comparison
plt.figure(figsize=(12, 5))

# Matplotlib
plt.subplot(1, 2, 1)
plt.plot(x, y, color="blue", linestyle="--", marker="o", markersize=5)
plt.title("Matplotlib Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)

# Seaborn
plt.subplot(1, 2, 2)
sns.lineplot(x="x", y="y", data=df, marker="o", linestyle="--", color="green")
plt.title("Seaborn Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)

plt.show()
````

### **2. Scatter Plot: Matplotlib vs. Seaborn**

````python
# Random Data
np.random.seed(10)
x = np.random.rand(50)
y = np.random.rand(50)

df = pd.DataFrame({"x": x, "y": y})

plt.figure(figsize=(12, 5))

# Matplotlib
plt.subplot(1, 2, 1)
plt.scatter(x, y, color="red", alpha=0.7, edgecolor="black")
plt.title("Matplotlib Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)

# Seaborn
plt.subplot(1, 2, 2)
sns.scatterplot(x="x", y="y", data=df, color="blue", edgecolor="black")
plt.title("Seaborn Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)

plt.show()
````

### **3. Bar Chart: Matplotlib vs. Seaborn**

````python
import warnings
warnings.filterwarnings("ignore")


# Data
categories = ["A", "B", "C", "D", "E"]
values = [10, 20, 15, 25, 30]
df = pd.DataFrame({"Category": categories, "Value": values})

plt.figure(figsize=(12, 5))

# Matplotlib
plt.subplot(1, 2, 1)
plt.bar(categories, values, color="green", alpha=0.7)
plt.title("Matplotlib Bar Chart")
plt.xlabel("Category")
plt.ylabel("Value")

# Seaborn
plt.subplot(1, 2, 2)
sns.barplot(x="Category", y="Value", data=df, palette="Greens")
plt.title("Seaborn Bar Chart")
plt.xlabel("Category")
plt.ylabel("Value")

plt.show()
````

### **4. Histogram: Matplotlib vs. Seaborn**

````python
# Random Data
data = np.random.randn(1000)

plt.figure(figsize=(12, 5))

# Matplotlib
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color="blue", edgecolor="black", alpha=0.7)
plt.title("Matplotlib Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Seaborn
plt.subplot(1, 2, 2)
sns.histplot(data, bins=30, kde=True, color="green")
plt.title("Seaborn Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
````

### **5. Pie Chart: Matplotlib vs. Seaborn**
Seaborn **does not support pie charts**, so we use **Matplotlib for both**.

````python
# Data
sizes = [25, 35, 20, 20]
labels = ["A", "B", "C", "D"]
colors = ["lightblue", "lightgreen", "orange", "pink"]

plt.figure(figsize=(12, 5))

# Matplotlib Pie Chart
plt.subplot(1, 2, 1)
plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%", startangle=90)
plt.title("Matplotlib Pie Chart")

# Matplotlib Styled Pie Chart (Seaborn alternative)
plt.subplot(1, 2, 2)
plt.pie(sizes, labels=labels, colors=sns.color_palette("pastel"), autopct="%1.1f%%", startangle=90)
plt.title("Styled Pie Chart using Matplotlib")

plt.show()
````
